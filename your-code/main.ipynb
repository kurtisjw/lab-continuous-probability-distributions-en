To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
import numpy as np
!pip install scipy
from scipy.stats import uniform

# Define the range
lower_bound = 10
upper_bound = 20

# Define the number of random numbers you want to generate
num_samples = 5

# Generate uniform random numbers
random_numbers = uniform.rvs(loc=lower_bound, scale=upper_bound-lower_bound, size=num_samples)

print("Generated random numbers:", random_numbers)
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
def generate_uniform_random_numbers(bottom, ceiling, count):
    """
    Generates an array of uniformly distributed random numbers.

    Parameters:
    bottom (float): The lower boundary of the generated numbers.
    ceiling (float): The upper boundary of the generated numbers.
    count (int): The number of random numbers to generate.

    Returns:
    numpy.ndarray: An array of uniformly distributed random numbers.
    """
    # Generate uniform random numbers
    random_numbers = uniform.rvs(loc=bottom, scale=ceiling-bottom, size=count)
    
    return random_numbers
```

```python
# Call the function with the specified parameters
random_numbers_1 = generate_uniform_random_numbers(10, 15, 100)
random_numbers_2 = generate_uniform_random_numbers(10, 60, 1000)
```

```python
# Plotting the histograms
def plot_histogram(data, bins, title):
    plt.hist(data, bins=bins, edgecolor='black')
    plt.title(title)
    plt.xlabel('Value')
    plt.ylabel('Count')
    plt.grid(True)
    plt.show()

# Plot the histogram for the first set of random numbers
plot_histogram(random_numbers_1, bins=10, title='Uniform Distribution (10, 15) with 100 samples')

# Plot the histogram for the second set of random numbers
plot_histogram(random_numbers_2, bins=10, title='Uniform Distribution (10, 60) with 1000 samples')
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm
```

```python
def generate_normal_random_numbers(mean, stddev, count):
    """
    Generates an array of normally distributed random numbers.

    Parameters:
    mean (float): The mean (average) of the distribution.
    stddev (float): The standard deviation of the distribution.
    count (int): The number of random numbers to generate.

    Returns:
    numpy.ndarray: An array of normally distributed random numbers.
    """
    random_numbers = norm.rvs(loc=mean, scale=stddev, size=count)
    return random_numbers

def plot_histogram(data, bins, title):
    plt.hist(data, bins=bins, edgecolor='black')
    plt.title(title)
    plt.xlabel('Value')
    plt.ylabel('Count')
    plt.grid(True)
    plt.show()

# Generate the random numbers
normal_random_numbers_1 = generate_normal_random_numbers(10, 1, 1000)
normal_random_numbers_2 = generate_normal_random_numbers(10, 50, 1000)

# Plot the histograms
plot_histogram(normal_random_numbers_1, bins=10, title='Normal Distribution (mean=10, stddev=1) with 1000 samples')
plot_histogram(normal_random_numbers_2, bins=10, title='Normal Distribution (mean=10, stddev=50) with 1000 samples')
```
